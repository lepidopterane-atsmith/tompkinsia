```python
results = pd.DataFrame({'Form': pd.Series([]), 'Lemma': pd.Series([]), 'Part of Speech': pd.Series([]),'Case': pd.Series([]), 'Number': pd.Series([]), 'Tense': pd.Series([]),
                        'Mood': pd.Series([]), 'Voice': pd.Series([]), 'Relation': pd.Series([]), 'Sentence ID': pd.Series([]), 'Word ID': pd.Series([]), 
                        'Document URN': pd.Series([])})

for i in all_urns:
    try:
        query_engine = create_engine_from_files(urns=[i])
        results = pd.concat([results, word_search(query_engine, "[form=σὺ] + δέ" )])
    except:
        print("oops")

results
```

Wow! That's a lot of results. Let's do some data wrangling to get a feel for how many of these results each author here is responsible for. Document URNs are individual identifiers for documents in the Greek lit corpus. The first four numbers correspond with an author, and the last four numbers correspond with a specific work. URNs will help us count these by author. 

```python
results['author'] = results['Document URN'].str.partition("-")[0]
real_results = results[results['Lemma'] == 'δέ']

authors = results['author'].unique()
hits = []

for a in authors:
    hit = len(real_results[real_results['author'] == a])
    hits.append(hit)

author_series = pd.Series(authors)
hits_series = pd.Series(hits)
author_data = pd.DataFrame({'author' : author_series, 'instances' : hits_series})

author_data
```

That's more like it. Let's see if we can use that list of unique authors to get a list of every URN they're responsible for.

```python
author_urns = []

for u in all_urns:
    for a in authors:
        if a in u:
            author_urns.append(u)
```

We want to see what percent of each author's works use our phrase, so we will borrow some code from another notebook to get token and punctuation counts to help us get a word count for our goal. 

```python
word_results = pd.DataFrame({'Form': pd.Series([]), 'Lemma': pd.Series([]), 'Part of Speech': pd.Series([]),'Case': pd.Series([]), 'Number': pd.Series([]), 'Tense': pd.Series([]),
                        'Mood': pd.Series([]), 'Voice': pd.Series([]), 'Relation': pd.Series([]), 'Sentence ID': pd.Series([]), 'Word ID': pd.Series([]), 
                        'Document URN': pd.Series([])})

all_words = []

for i in author_urns:
    query_engine = create_engine_from_files(urns=[i])
    punct, tokens = len(word_search(query_engine, ":punctuation" )) , len(word_search(query_engine, ":neighbor" )) 
    # second statement grabs all tokens including punctuation
    result = tokens - punct
    all_words.append(result)
```

We can smash this again from URNs to authors, and then it's time to prepare our results.

```python
word_data = pd.DataFrame({'author urns': pd.Series(author_urns), 'word count': pd.Series(all_words)})
word_data['author'] = word_data['author urns'].str.partition("-")[0]

word_counts = []

for a in authors:
    hit = word_data[word_data['author'] == a]['word count'].sum()
    word_counts.append(hit)

wc_series = pd.Series(word_counts)
author_data['word count'] = wc_series

# results time!

author_data['percent saturation']= author_data['instances'] * 100 / author_data['word count']
author_data.sort_values('percent saturation', ascending=False)[:20]
```

Because of weird sample size, our frontrunners here are actually 0533 (Callimachus), 0019 (Aristophanes), and 0085 (Aeschylus). However, let's see what we happens when sorting by most instances.

```python
author_data.sort_values('instances', ascending=False)[:20]
```

Sure enough, Homer (0012) is in third, alongside the likes of Lucian of Samosata (0062) and Plato (0059). However, as fans of lexical dispersion will comment, high volume of a feature does not always correlate with a strong consistency throughout one's work to use it.
